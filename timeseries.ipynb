```python
import numpy as np
import pandas as pd
import datetime
from functools import reduce
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
import math
```

```python
df = pd.read_csv("Updated_Retailer_Liquadation.csv")
df.head()

df.head()
```

```python
df.columns
```

```python
lst = []
lst = df['state']
df = df.loc[df['type']=="sales"] 
y = df[["available_stock"]]
```

```python
df.drop(columns=['id', 'product_id',
        'sold_date', 'pending_stock', 'available_stock',
       'type', 'end_date'],inplace = True)
df.dropna()
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df["retailer_id"].max()
        min_value = df["retailer_id"].min()
        result["retailer_id"] = (df["retailer_id"] - min_value) / (max_value - min_value)
    return result
```

```python
df = normalize(df)
```

```python
df["year"]= pd.DatetimeIndex(df['received_date']).year
df["month"]= pd.DatetimeIndex(df['received_date']).month
df["day"]= pd.DatetimeIndex(df['received_date']).day
df.drop(columns = ["received_date"])
```

```python
x = df.copy()
y.head()
y["available_stock"]= y["available_stock"].astype(np.float32)
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

state_dum = pd.get_dummies(df.state)
x = pd.concat([x,state_dum ], axis=1)
x.drop(columns = ["state"],inplace = True)
x.drop(columns = ["received_date"],inplace = True)
x.head()
x["retailer_id"]= x["retailer_id"].astype(np.float32)
x["sold_stock"]= x["sold_stock"].astype(np.float32)
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
y_test.dtypes
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 100, random_state = 40)
# Train the model on training data
rf.fit(X_train, y_train.values.ravel())
```
